```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("47176.txt", sep='\t', encoding='latin-1')
df.head(20)
```

```python
df[df["TraitID"].isna()]
```

```python
df_trait_id = df[df["TraitID"].notna()]
```

```python
df_trait_id.info()
```

```python
df_trait_id_non_nan = df_trait_id[df_trait_id['StdValue'].notna()]
```

```python
df_trait_id_non_nan.loc[:, "AccSpeciesName"] = (
    df_trait_id_non_nan["AccSpeciesName"].str.lower()
)
df_trait_id_non_nan.groupby("AccSpeciesName").size()
```

```python
df_grouped = df_trait_id_non_nan.groupby("AccSpeciesName", as_index=False).agg(
    mean_height_min_m=("StdValue", "mean"),
    median_height_min_m=("StdValue", "median"),
    references=("Reference", lambda x: "; ".join(x.dropna().unique()))
)
df_grouped
```

```python
df_trait_id_non_nan.to_csv("cleaned_plant_height.csv", index=False)
```
