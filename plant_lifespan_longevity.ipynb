```python
import pandas as pd
```

```python
df = pd.read_csv("47180.txt", sep='\t', encoding='latin-1')
df.head()
```

```python
df_trait_id = df[df["TraitID"]== 59]
df_trait_id_non_nan = df_trait_id[df_trait_id['StdValue'].notna()]
df_trait_id_non_nan.count()
```

```python
df_trait_id_non_nan.head()
```

```python
df_trait_id_non_nan["UnitName"].value_counts()
```

```python
df_trait_id_non_nan.nunique()
```

```python
df_trait_id_non_nan.loc[:, "SpeciesName"] = (
    df_trait_id_non_nan["SpeciesName"].str.lower()
)
df_trait_id_non_nan.loc[:, "AccSpeciesName"] = (
    df_trait_id_non_nan["AccSpeciesName"].str.lower()
)
```

```python
df_mismatch = df_trait_id_non_nan[
    df_trait_id_non_nan["SpeciesName"] != df_trait_id_non_nan["AccSpeciesName"]
]
df_mismatch
```

```python
df_trait_id_non_nan.groupby("AccSpeciesName").size()
```

```python
df_grouped = df_trait_id_non_nan.groupby("AccSpeciesName", as_index=False).agg(
    mean_life_exp=("StdValue", "mean"),
    median_life_exp=("StdValue", "median"),
    references=("Reference", lambda x: "; ".join(x.dropna().unique()))
)
df_grouped
```

```python
df_trait_id_non_nan.to_csv("cleaned_plant_lifespan_longevity.csv", index=False)
```
