```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("47178.txt", sep='\t', encoding='latin-1')
df.head()
```

```python
df_trait_id = df[df['TraitID'].notna()]
df_trait_id.head()
```

```python
df_trait_id.info()
```

```python
# Difference in OrigValueStr and StdValue
df_trait_id[
    df_trait_id["OrigValueStr"].notna() &
    df_trait_id["StdValue"].isna()
]
```

```python
unit = df_trait_id["OrigValueStr"].str.extract(r"\b(cm|m|ft)\b")[0]
nums = df_trait_id["OrigValueStr"].str.findall(r"\d+(?:\.\d+)?")

def max_or_nan(x):
    if isinstance(x, list) and len(x) > 0:
        return max(map(float, x))
    return np.nan

max_value = nums.apply(max_or_nan)

conversion = {
    "m": 1.0,
    "cm": 0.01,
    "ft": 0.3048
}

converted = max_value * unit.map(conversion)

df_trait_id["StdValue"] = df_trait_id["StdValue"].fillna(converted)
df_trait_id[df_trait_id["OrigValueStr"] == "100 cm"]
```
